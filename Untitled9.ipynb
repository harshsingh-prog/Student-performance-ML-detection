```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_excel('FinalData.xlsx')
```

```python
dataset
```

```python
dataset.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
# split dataset
X = dataset.iloc[:, 4:5]
y = dataset.iloc[:, 5]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
#Feature Scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import math
math.sqrt(len(y_train))
```

```python
import math
math.sqrt(len(y_test))
```

```python
#define model
classifier = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
```

```python
#fit model
classifier.fit(X_train, y_train)
```

```python
#predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
f1 = f1_score(y_test, y_pred, average='macro')
print("F1 Score:", f1)
```

```python
f1 = f1_score(y_test, y_pred, average='micro')
print("F1 Score:", f1)
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1 Score:", f1)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Assume y_true and y_score are your true labels and predicted scores
y_true = [0, 0, 1, 1, 1]
y_score = [0.1, 0.4, 0.35, 0.8, 0.9]

# Calculate precision and recall for different thresholds
precision, recall, thresholds = precision_recall_curve(y_true, y_score)

# Plot the precision-recall curve
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()

```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# split dataset
X = dataset.iloc[:, 4:44]
y = dataset.iloc[:, 44]
scaler = MinMaxScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
# Perform PCA for dimensionality reduction (optional)
pca = PCA(n_components=10)
dataset_scaled = scaler.fit_transform(X_train, X_test)
dataset_pca = pca.fit_transform(dataset_scaled)
```

```python

# Train a Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)
```

```python

# Evaluate the model
accuracy = rfc.score(X_test, y_test)
print(f'Accuracy: {accuracy:.3f}')
```
